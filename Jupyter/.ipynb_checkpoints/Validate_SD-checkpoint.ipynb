```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_old = np.load("../data/simu_20000_0.1_90_140_train.npy")
data_new = np.load("../data/simu_10000_0.1_90_140_resonance_morlet.npy")
d21_old = np.load("../data/features_train.npy")[:,2]-np.load("../data/features_train.npy")[:,1]
d21_new = np.load("../data/D21_resonance.npy")
```
